```python
encoded_cleaned_test_df=pd.read_csv("/Users/amelchristy/Downloads/Amazon_R/encoded_cleaned_test.csv")
```

```python
# Create a new DataFrame to hold the combined unique mappings for all specified columns
all_combined_unique_df = pd.DataFrame()

# Loop over each column to create unique mappings and add them to the DataFrame
for col in ['Weather', 'Road_traffic_density', 'Vehicle_condition', 'Type_of_order', 'Type_of_vehicle','City']:
    # Combine the encoded and cleaned columns into a single column in the format "encoded:cleaned"
    combined_col = encoded_cleaned_test_df[col].astype(str) + ':' + cleaned_test_df[col].astype(str)
    # Get unique combinations and create a DataFrame
    unique_col_df = pd.DataFrame({col + '_combined': combined_col.unique()})
    # Concatenate to the main DataFrame
    all_combined_unique_df = pd.concat([all_combined_unique_df, unique_col_df], axis=1)

# Display the DataFrame with all unique mappings
all_combined_unique_df
```

```python
amazon =pd.read_csv("/Users/amelchristy/Downloads/Amazon_R/updated.csv")
```

```python
# Define the mappings based on your provided combined unique values
# Ensure the keys are integers if the original dataframe has integers
weather_mapping = {7: None, 6: 'Windy', 3: 'Stormy', 1: 'Fog', 4: 'Sunny', 0: 'Cloudy', 2: 'Sandstorms'}
traffic_mapping = {4: None, 1: 'Jam', 3: 'Medium', 2: 'Low', 0: 'High'}
vehicle_condition_mapping = {3: 'Poor', 0: 'Excellent', 1: 'Good', 2: 'Fair'}
type_of_order_mapping = {1: 'Drinks', 3: 'Snack', 2: 'Meal', 0: 'Buffet'}
type_of_vehicle_mapping = {1: 'Electric Scooter', 2: 'Motorcycle', 3: 'Scooter', 4: 'Bicycle'}
city_mapping = {0: 'Metropolitian', 1: 'Urban', 2: 'Semi-Urban'}



# Apply the mappings to the 'amazon' dataframe
amazon['Weatherconditions'] = amazon['Weatherconditions'].map(weather_mapping).fillna(amazon['Weatherconditions'])
amazon['Road_traffic_density'] = amazon['Road_traffic_density'].map(traffic_mapping).fillna(amazon['Road_traffic_density'])
amazon['Vehicle_condition'] = amazon['Vehicle_condition'].map(vehicle_condition_mapping).fillna(amazon['Vehicle_condition'])
amazon['Type_of_order'] = amazon['Type_of_order'].map(type_of_order_mapping).fillna(amazon['Type_of_order'])
amazon['Type_of_vehicle'] = amazon['Type_of_vehicle'].map(type_of_vehicle_mapping).fillna(amazon['Type_of_vehicle'])
amazon['City'] = amazon['City'].map(city_mapping).fillna(amazon['City'])

amazon.head()
```

```python
amazon.to_csv('.../amazon_updated.csv', index=False)
```
